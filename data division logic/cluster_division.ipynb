```python
from datasets import load_dataset
import json

# --- Configuration ---
JSONL_FILENAME = "miriad_4.4M.jsonl"
BATCH_SIZE = 10000  # Adjust based on Colab RAM. Smaller is safer.
# ---------------------

# 1. Load the dataset in streaming mode
ds = load_dataset("miriad/miriad-4.4M", split="train", streaming=True)

i = 0
rows_processed = 0

print(f"Starting to write dataset to {JSONL_FILENAME}...")

try:
    with open(JSONL_FILENAME, 'w', encoding='utf-8') as f:
        for row in ds:
            # 2. Write the row as a JSON string followed by a newline (JSONL format)
            # The 'row' object is a Python dictionary
            f.write(json.dumps(row) + '\n')

            rows_processed += 1

            if rows_processed % BATCH_SIZE == 0:
                i += 1
                print(f"Processed and wrote {rows_processed} rows (Batch {i})...")

except Exception as e:
    print(f"\n❌ An error occurred during writing: {e}")

print(f"\n✅ Finished processing. Total rows saved: {rows_processed} to {JSONL_FILENAME}")
```

```python
import json

INPUT_FILE = "miriad_4.4M.jsonl"
OUTPUT_FILE = "miriad_neurology.jsonl"

# --- Multiple specialties here ---
TARGET_SPECIALTIES = [
   "Neurology",
   "Neurosurgery"
]  # <--- Add/remove specialties as needed
# --------------------------------

count_in = 0
count_out = 0

print(f"Starting filtering for specialties: {TARGET_SPECIALTIES}")

with open(INPUT_FILE, "r", encoding="utf-8") as infile, \
     open(OUTPUT_FILE, "w", encoding="utf-8") as outfile:

    for line in infile:
        count_in += 1
        row = json.loads(line)

        # Check if the row's specialty is in the list
        if row.get("specialty") in TARGET_SPECIALTIES:
            outfile.write(json.dumps(row) + "\n")
            count_out += 1

        # Show progress every 100k rows
        if count_in % 100000 == 0:
            print(f"Processed {count_in} lines... Kept {count_out}")

print("\nDone!")
print(f"Total rows scanned: {count_in}")
print(f"Rows kept for specialties {TARGET_SPECIALTIES}: {count_out}")
print(f"Saved to: {OUTPUT_FILE}")
```

```python
import json

INPUT_FILE = "miriad_neurology.jsonl"
OUTPUT_FILE = "miriad_neurology.json"

print("Converting JSONL → JSON...")

with open(INPUT_FILE, "r", encoding="utf-8") as infile, \
     open(OUTPUT_FILE, "w", encoding="utf-8") as outfile:

    outfile.write("[\n")   # start of JSON array

    first = True
    for line in infile:
        if not first:
            outfile.write(",\n")  # comma BEFORE every item except the first
        first = False

        obj = json.loads(line)
        outfile.write(json.dumps(obj, ensure_ascii=False))

    outfile.write("\n]")   # end of JSON array

print("Done! Saved as:", OUTPUT_FILE)
```
